```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Load the dataset
df = pd.read_csv('Details _ comments YouTube video.csv', encoding='ISO-8859-1')

# Check if "Comment_Datail" column exists
if "Comment_Datail" in df.columns:
    # Preprocessing function
    def preprocess_text(text):
        # Remove special characters and numbers
        text = re.sub(r'[^A-Za-z\s]', '', text)
        
        # Convert to lowercase
        text = text.lower()
        
        # Tokenize the text
        words = word_tokenize(text)
        
        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        filtered_text = [word for word in words if word not in stop_words]
        
        return ' '.join(filtered_text)

    # Apply preprocessing to the "Comment_Datail" column
    df['Cleaned_Comment_Datail'] = df['Comment_Datail'].apply(preprocess_text)

    # Save the cleaned data to a new CSV file
    df.to_csv('Cleaned_Details_comments_YouTube_video.csv', index=False)
else:
    print('Column "Comment_Datail" does not exist in the CSV file.')
```

```python
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Combine all text in the Cleaned_Comment_Datail column into one large text corpus
all_text = ' '.join(df['Cleaned_Comment_Datail'].dropna())

# Tokenize the combined text
word_tokens = word_tokenize(all_text)

# Count word frequencies
word_freq = Counter(word_tokens)

# Display the 10 most common words
common_words = word_freq.most_common(10)
print(common_words)

# Plot the most common words
words, counts = zip(*common_words)
plt.figure(figsize=(10,6))
plt.bar(words, counts)
plt.title('Top 10 Most Frequent Words')
plt.show()
```

```python
# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Display word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Cleaned Comments')
plt.show()
```

```python
from nltk import ngrams

# Function to get the most common n-grams (in this case, bigrams)
def get_ngrams(text, n=2):
    tokenized = word_tokenize(text)
    n_grams = ngrams(tokenized, n)
    return Counter(n_grams)

# Get the most common bigrams
bigram_freq = get_ngrams(all_text, n=2)
common_bigrams = bigram_freq.most_common(10)
print(common_bigrams)

# Plot the most common bigrams
bigrams, counts = zip(*common_bigrams)
bigrams = [' '.join(bigram) for bigram in bigrams]  # Join bigrams into readable format
plt.figure(figsize=(10,6))
plt.bar(bigrams, counts)
plt.title('Top 10 Most Frequent Bigrams')
plt.xticks(rotation=45)
plt.show()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to analyze sentiment using VADER
def vader_sentiment_analysis(text):
    sentiment_dict = analyzer.polarity_scores(text)
    
    # Classify as positive, negative, or neutral based on compound score
    if sentiment_dict['compound'] >= 0.05:
        return 'Positive'
    elif sentiment_dict['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply VADER sentiment analysis to the "Cleaned_Comment_Datail" column
df['VADER_Sentiment'] = df['Cleaned_Comment_Datail'].apply(lambda x: vader_sentiment_analysis(str(x)))

# View the results
df[['Cleaned_Comment_Datail', 'VADER_Sentiment']].head()

# Save the results to a new CSV
df.to_csv('Sentiment_Analysis_VADER_Results.csv', index=False)
```

```python
from textblob import TextBlob

# Function to analyze sentiment using TextBlob
def textblob_sentiment_analysis(text):
    # Get the polarity score (-1 to 1)
    polarity = TextBlob(text).sentiment.polarity
    
    # Classify as positive, negative, or neutral
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply TextBlob sentiment analysis to the "Cleaned_Comment_Datail" column
df['TextBlob_Sentiment'] = df['Cleaned_Comment_Datail'].apply(lambda x: textblob_sentiment_analysis(str(x)))

# View the results
df[['Cleaned_Comment_Datail', 'TextBlob_Sentiment']].head()

# Save the results to a new CSV
df.to_csv('Sentiment_Analysis_TextBlob_Results.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Load the labeled dataset
# Assuming that 'Cleaned_Comment_Datail' contains the preprocessed text
# and 'Sentiment_Label' contains the sentiment (Positive/Negative/Neutral)

# Example: df['Sentiment_Label'] should be filled with sentiment labels
# You can use VADER or TextBlob sentiment results for initial labeling if needed.

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Cleaned_Comment_Datail'], df['VADER_Sentiment'], test_size=0.2, random_state=42)

# Convert the text data into numerical form using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

### 1. Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(X_train_tfidf, y_train)
y_pred_lr = lr_model.predict(X_test_tfidf)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))

### 2. Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
y_pred_nb = nb_model.predict(X_test_tfidf)
print("Naive Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))

### 3. Support Vector Machine (SVM)
svm_model = SVC()
svm_model.fit(X_train_tfidf, y_train)
y_pred_svm = svm_model.predict(X_test_tfidf)
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Vectorize the comments
vectorizer = CountVectorizer(max_df=0.9, min_df=2, stop_words='english', max_features=1000)
X_aspects = vectorizer.fit_transform(df['Cleaned_Comment_Datail'])

# Apply LDA to extract topics
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda_aspects = lda.fit_transform(X_aspects)

# Function to display the topics and their top words
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}: ", [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]])

# Display the top 10 words in each topic (aspect)
display_topics(lda, vectorizer.get_feature_names_out(), 10)
```

```python
df['Aspect'] = lda_aspects.argmax(axis=1)
```
